```python
 #Principal component analysis

#Considering only the 11 cols and converting it into array
cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']
x = df.loc[:,cat].values
x
from sklearn.decomposition import PCA
from sklearn import preprocessing

pca_data = preprocessing.scale(x)

pca = PCA(n_components=11)
pc = pca.fit_transform(x)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
pf = pd.DataFrame(data = pc, columns = names)
pf
```

```python
#Proportion of Variance (from PC1 to PC11)
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
# correlation coefficient between original variables and the component
#Considering only first 11 attributes
df_eleven = df.loc[:,cat]
df_eleven
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df_eleven.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
#Correlation matrix plot for loadings 
plt.rcParams['figure.figsize'] = (20,15)
ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
plt.show()
```

```python
#Scree plot (Elbow test)- PCA
from bioinfokit.visuz import cluster
cluster.screeplot(obj=[pc_list, pca.explained_variance_ratio_],show=True,dim=(10,5))
```

```python
# get PC scores
pca_scores = PCA().fit_transform(x)

# get 2D biplot
cluster.biplot(cscore=pca_scores, loadings=loadings, labels=df.columns.values, var1=round(pca.explained_variance_ratio_[0]*100, 2),
    var2=round(pca.explained_variance_ratio_[1]*100, 2),show=True,dim=(10,5))
```

# Step 6: Profiling Segments

```python
from scipy.cluster import hierarchy
from scipy.spatial.distance import pdist

MD_x=X

# Assuming MD.x contains the data
distance_matrix = pdist(MD_x.T)  # Calculate the pairwise distance matrix
MD_vclust = hierarchy.linkage(distance_matrix, method='complete')  # Perform hierarchical clustering
```

```python
np.random.seed(1234)

cluster_range = range(2, 9)
nrep = 10

# Perform flexible clustering for a range of cluster numbers
results = []
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, n_init=nrep)
    kmeans.fit(MD_x)
    results.append(kmeans)

# Relabel the clusters
relabel_results = []
for kmeans in results:
    labels = kmeans.labels_
    unique_labels = np.unique(labels)
    relabeled_labels = np.arange(len(unique_labels))
    relabeled_labels_map = dict(zip(unique_labels, relabeled_labels))
    relabeled_labels = np.array([relabeled_labels_map[label] for label in labels])
    relabel_results.append(relabeled_labels)
```

```python
MD_k4 = relabel_results[2] 
```

```python
# Assuming MD_k4 is the data for the bar chart
MD_vclust_order = np.flip(MD_vclust)  # Reverse the order obtained from MD.vclust$order

# Create a color map for shading the bars
colors = ['lightblue' if i in MD_vclust_order else 'gray' for i in range(len(MD_k4))]

# Plot the bar chart
plt.bar(range(len(MD_k4)), MD_k4, color=colors)

plt.xlabel('Segment')
plt.ylabel('Percentage')
plt.title('Segment Profile Plot')

plt.show()
```
